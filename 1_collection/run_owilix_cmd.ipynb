```python
import subprocess
from datetime import datetime, timedelta

def run_cmd_with_dates_and_languages(base_command, start_date, days, languages, conda_env):
    date_list = [start_date + timedelta(days=i) for i in range(days)]

    processes = []
    for i, date in enumerate(date_list):
        for language in languages:
            command = base_command.replace("all:2024-04-09", f"all:{date.strftime('%Y-%m-%d')}")
            command = command.replace("language=vie", f"language={language}")
            full_command = f"{command}"
            print(f"Starting command ({i + 1}/{days}) for language {language}: {full_command}")
            process = subprocess.Popen(full_command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, executable="/bin/bash")
            processes.append((process, date, language))

    # Collect results
    for i, (process, date, language) in enumerate(processes):
        stdout, stderr = process.communicate()
        if process.returncode == 0:
            print(f"Command output for {date.strftime('%Y-%m-%d')} - {language} ({i + 1}): {stdout}")
        else:
            print(f"Error for {date.strftime('%Y-%m-%d')} - {language} ({i + 1}): {stderr}")

# Define the base command
base_cmd = (
    "owilix --remotes it4i query less --remote all:2024-04-09 "
    "limit=500 files=\"**/language=vie/*parquet\" as_json=True"
)

# Specify the start date, number of days, and languages
start_date = datetime.strptime("2024-01-01", "%Y-%m-%d")
days_to_run = 20
languages = ["vie", "deu","eng"]
conda_env = "ros_env"

# Run the command for multiple dates and languages in parallel
run_cmd_with_dates_and_languages(base_cmd, start_date, days_to_run, languages, conda_env)
```
